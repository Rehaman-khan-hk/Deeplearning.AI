```python
from transformers.utils import logging

logging.set_verbosity_error()
```

```python
import os
from transformers import pipeline

# Define model path
local_model_path = "./models/kakao-enterprise/vits-ljs"

# Check if the model directory exists
if not os.path.exists(local_model_path):
    print("Error: Model directory not found. Using Hugging Face model instead.")
    local_model_path = "kakao-enterprise/vits-ljs"  # Use online model

# Load the text-to-speech pipeline
try:
    narrator = pipeline("text-to-speech", model=local_model_path)
    print("Model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {e}")
```
